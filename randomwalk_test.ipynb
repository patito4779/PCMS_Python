```python
import numpy as np
import random
from matplotlib.pyplot import *
import sys
```

```python
particles = 10000
steps = 1000

def random_walk(particles, steps):
    north = 1
    south = 2
    east = 3
    west = 4
    xparticle = np.zeros(particles)
    yparticle = np.zeros(particles)
    for step in range(steps):
        
        for i in range(particles):
            direction = random.randint(1, 4)
            if direction == north:
                yparticle[i] += 1
            elif direction == south:
                yparticle[i] -= 1
            elif direction == east:
                xparticle[i] += 1
            else:
                xparticle[i] -= 1
    return xparticle, yparticle

x, y = random_walk(particles, steps)

plot(x, y, "bo")
```
