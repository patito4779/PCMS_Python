```python
# Now let us implement the ball function
from math import *
import numpy as np
from matplotlib.pylab import *
```

```python
g = 9.8
def ball(v0, theta, y0, x):
    v0 = v0 /3.6
    theta = radians(theta)
    
    return x * tan(theta) - 1./(2*v0**2)*g * x**2/(cos(theta)**2) + y0

x = np.linspace(0, 2, 100)



```

```python
plot(x, ball(15, 60, 1, x))
```

```python

for v0 in range(10, 60, 10):
    y = ball(v0, 60, 1, x)
    plot(x,y)
show()
    
```

```python
for theta in range(35, 60, 5):
    y = ball(15, theta, 1, x)

    plot(x, y)
show()
```
