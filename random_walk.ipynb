```python
import numpy as np
import random
import matplotlib.pyplot as plt
import sys

# define random walk function
def random_walk_2d(particles, steps):
  #setup arrays
  xpositions = np.zeros(particles)
  ypositions = np.zeros(particles)
  #define directions
  NORTH = 1
  EAST  = 2
  SOUTH = 3
  WEST  = 4
  #loop steps
  for step in range(steps):
    #loop particles
    for i in range(particles):
      #random number
      direction = random.randint(1,4)
      #displacement
      if direction == NORTH:
        ypositions[i] += 1
      elif direction == EAST:
        xpositions[i] += 1
      elif direction == SOUTH:
        ypositions[i] -= 1
      elif direction == WEST:
          xpositions[i] -= 1
  return xpositions, ypositions

# set N(particles) and N(steps)
particles = 10000
steps     = 1000

#call random walker
x,y = random_walk_2d(particles, steps)

#plot result
plt.plot(x,y,"bo")
plt.show()
```
