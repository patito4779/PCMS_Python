```python
from math import *
from matplotlib.pylab import *
import numpy as np
```

```python
def f(t):
    return t**2*exp(-t**2)
```

```python
t = np.linspace(0, 3, 51)
y = np.zeros(len(t))
```

```python
print(len(t))
print(t)
print(y)
```

```python
for i in range(len(t)):
    y[i] = f(t[i])
plot(t,y)
xlabel("t")
ylabel("y")
title("my first matplotlib ")
```
